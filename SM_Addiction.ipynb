```python
import matplotlib.pyplot as plt
gender_counts = data.Gender.value_counts()

labels = gender_counts.index.tolist()
values = gender_counts.values.tolist()

plt.bar(labels, values, color=['lightblue', 'salmon'])

for i, val in enumerate(values):
    plt.text(i, val + 0.5, str(val), ha='center', fontsize=10)


plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#In map function-we pass dictionary
data['Affects_Academic_Performance_Numeric'] = data['Affects_Academic_Performance'].map({'Yes': 1, 'No': 0})
print(data[['Affects_Academic_Performance', 'Affects_Academic_Performance_Numeric']].head())
print(data['Affects_Academic_Performance_Numeric'].dtype)
data = data.drop('Affects_Academic_Performance', axis=1)
data.dtypes
```

```python
data['Relationship_Status_numeric']= data['Relationship_Status'].map({'Single':1,'In Relationship':2,'Complicated':3})
data[['Relationship_Status','Relationship_Status']].head()
data['Relationship_Status_numeric'].dtypes
```

```python
data_corr=data[['Avg_Daily_Usage_Hours','Affects_Academic_Performance_Numeric','Sleep_Hours_Per_Night','Mental_Health_Score','Relationship_Status_numeric']].corr()
data_corr
```

```python
corr_matrix = data_corr
plt.figure(figsize=(6,8))
sns.heatmap(corr_matrix, annot=True, cmap = 'inferno')
plt.show()
```

```python
index = np.arange(len(data))
data.groupby('Gender')['Addicted_Score'].mean().plot(kind='bar',color='skyblue',edgecolor='black',width=0.5,figsize=(5,5))
plt.title('Gender vs Addiction Score')
plt.xlabel('Gender')
plt.ylabel('Addicted Score')
plt.show()
```

```python
avg_usage_by_age = data.groupby('Age')['Avg_Daily_Usage_Hours'].mean()
plt.plot(avg_usage_by_age.index, avg_usage_by_age.values, marker='o', color='skyblue',markerfacecolor='darkblue')
plt.title('Average Daily Usage Hours by Age')
plt.xlabel('Age')
plt.ylabel('Average Daily Usage Hours')
plt.grid(True)
plt.tight_layout()
plt.show()
```
